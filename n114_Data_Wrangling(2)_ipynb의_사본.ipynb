# 1. customer 테이블의 데이터 이슈 해결하기 
- customer 데이터셋을 불러옵니다. 
- 마지막 row를 제거합니다. 
- `customer_id`, `store_id`, `address_id`, `active` 컬럼의 데이터 타입을 int로 변환합니다. 
- `name`에서 처음 이름은 `first_name` 컬럼에, 다음 이름은 `last_name` 컬럼에 저장합니다. 이후 `name` 컬럼은 제거합니다.
- `customer_id` 컬럼에 대하여 오름차순으로 정렬합니다. 이후 `.reset_index()`를 사용하여 인덱스를 재설정 해 주세요. 

```python
import pandas as pd
```

```python
customer = pd.read_csv('/content/drive/My Drive/customer', index_col=0)
customer
```

```python
customer.drop(599, inplace=True)
customer.tail(3)
```

```python
 customer.info()
```

```python
columns = ['customer_id', 'store_id', 'address_id', 'active']
 
customer[columns] = customer[columns].astype(int)
customer.dtypes
```

```python
customer[['first_name', 'last_name']] = customer['name'].str.split(',', 1, expand=True)
```

```python
customer.drop('name', axis=1, inplace=True)
```

```python
customer.sort_values('customer_id', inplace=True)

customer.reset_index(drop=True, inplace=True)
customer
```

# 3. payment 테이블의 데이터 이슈 해결하기 
- payment 데이터셋을 불러옵니다. 
- 마지막 row를 제거합니다. 
- `payment_id`, `customer_id`, `staff_id`, `rental_id` 컬럼의 데이터 타입을 int로 변환합니다. 
- `amount` 컬럼의 '$'를 제거한 후 데이터 타입을 float으로 변환합니다. 
- `customer_id`와 `rental_id` 컬럼에 대하여 오름차순으로 정렬합니다. 

```python
payment = pd.read_csv('/content/drive/My Drive/payment', index_col=0)
payment
```

```python
payment.drop(14596, inplace=True)
payment.tail(3)
```

```python
 payment.info()
```

```python
columns = ['payment_id', 'customer_id', 'staff_id', 'rental_id']

payment[columns] = payment[columns].astype(int)
```

```python
payment['amount'] = payment['amount'].str.replace('$', '', regex=True).astype('float')
payment
```

```python
payment.dtypes
```

```python
payment.sort_values(['customer_id', 'rental_id'], inplace=True)
payment
```

# 4. customer, rental, payment 테이블 합치기 
3개의 테이블을 모두 사용하여 분석을 진행하는 것은 비효율적으로 보입니다.
- rental 테이블과 payment 테이블의 `staff_id` 컬럼을 각각 제거합니다.  
- customer 테이블과 rental 테이블의 `last_update` 컬럼을 각각 제거합니다. 
- rental 테이블과 payment 테이블을 `customer_id`와 `rental_id`에 대하여 `.merge()`를 사용하여 합치세요. 변수 이름은 `df1`으로 합니다. 
- `df1`과 customer 테이블을 `customer_id`에 대하여 `.merge()`를 사용하여 합치세요. 변수 이름은 `df2`로 합니다. 

```python
rental.drop('staff_id', axis=1, inplace=True)
payment.drop('staff_id', axis=1, inplace=True)
```

```python
customer.drop('last_update', axis=1, inplace=True)
rental.drop('last_update', axis=1, inplace=True)
```

```python
df1 = pd.merge(rental, payment)
df1
```

```python
df2 = pd.merge(df1, customer, on=['customer_id'])
df2
```
